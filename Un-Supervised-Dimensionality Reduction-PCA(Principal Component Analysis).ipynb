### (B).Dimensionality Reduction-PCA(Principal Component Analysis)
#### It is a linear dimensionality reduction method which uses singular value decomposition of the data and keeps only the most significant singular vectors to project the data to a lower dimensional space.

```python
# Import required Libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# define no of samples
n_samples=50
# define random state value to inialize the center
random_state=20
#define the no. of feature as 5
X,y=make_blobs(n_samples=n_samples,n_features=20,random_state=None)
#define number of cluster to be formed as 5 and in random state and fit features into the model
predict_y=KMeans(n_clusters=5,random_state=random_state).fit_predict(X)
predict_y
```

```python
# Define pca estimator with number of reduce components
# HERE WE REDUCE THE FETURE
pca=PCA(n_components=20)
```

```python
#fit the dta into PCA estimator
# HERE WE REDUCE THE FETURE from 20 to 12
pca.fit(X)
print(pca.explained_variance_ratio_)
```

```python
# print first pca component
first_pca=pca.components_[11]
print(first_pca)
```

```python
# transport the fitted Data using transform method
pca_reduced=pca.transform(X)
```

```python
#view the reduce shape (lower dimension of feature from 15 to 10)
pca_reduced.shape
```

```python
# first use KNN classifier method - import it from sklearn
from sklearn.neighbors import KNeighborsClassifier
```

```python
# instantiate the knn estimator
knn = KNeighborsClassifier()
```

```python
knn.fit(X,y)
```

```python
y_predict=knn.predict(X)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y,y_predict))
```

```python
## Evaluate Model Performance
from sklearn.metrics import confusion_matrix
cm=pd.DataFrame(confusion_matrix(y,y_predict))
cm['Total']=np.sum(cm,axis=1)
cm=cm.append(np.sum(cm,axis=0),ignore_index=True)
cm.columns=["Predicted 1","Predicted 2","Predicted 3","Total"]
cm=cm.set_index([["Actual 1","Actual 2","Actual 3","Total"]])
print(cm)
```
