```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
digits_dataset=pd.DataFrame(digits.data)
digits_dataset
```

```python
digits.target
```

```python
digits.target_names
```

```python
digits_dataset["target"]=pd.DataFrame(digits.target)
```

```python
digits_dataset.head(n=20)
```

### Train the model and prediction



```python
from sklearn.model_selection import train_test_split
```

```python
X=digits_dataset.drop(['target'], axis='columns')
X
```

```python
y=digits.target
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
iris_dataset.target_names
```

```python
df_iris_dataset=pd.DataFrame(iris_dataset.data,columns=iris_dataset.feature_names)
df_iris_dataset
```

```python
df_iris_dataset["target"]=pd.DataFrame(iris_dataset.target)
df_iris_dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_iris_dataset.drop(['target'], axis='columns')
X
```

```python
y=df_iris_dataset.target
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
