## Histogram,Scatter Plot & Heat Map by Matplotlib & Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
# load  boston data set
boston_real_estate_data=load_boston()
```

```python
# view boston data set
print(boston_real_estate_data.DESCR)
```

## Histogram

```python
# define x Axis
x_axis=boston_real_estate_data.data
# define x Axis
y_axis=boston_real_estate_data.target
```

```python
# plot histogram
style.use("ggplot")
plt.figure(figsize=(7,7))
plt.hist(y_axis,bins=60)
plt.xlabel("Pice in 1000 USD")
plt.ylabel("No of houses.")
plt.show()
```

## Scatter Plot

```python
# plot scatterplot
style.use("ggplot")
plt.figure(figsize=(6,6))
plt.title("Boston Real Estate Price")
plt.scatter(boston_real_estate_data.data[:,6],boston_real_estate_data.target)
plt.xlabel("Price in 1000 USD")
plt.ylabel("No of houses.")
plt.show()
```

## Heat Map

```python
# Load Flight Data from sns Data set
flight_data=sns.load_dataset("flights")
```

```python
# view top 5 records
flight_data.head()
```

```python
## use pivot method to rearrange the data set
flight_data=flight_data.pivot("month","year","passengers")
```

```python
flight_data
```

```python
## Heat map for Flight
sns.heatmap(flight_data,annot=True,fmt="d", linewidths=.5,cmap="bwr")
```

## Pie Chart

```python
Job_data=[230,300,100,61,105,250]

Department="IT","Marketing","Finanace","HR","Admin","Operations"

colors=["Orange","Green","Purple","Cyan","Royalblue","Magenta"]

explode=(0.05,0.08,0.06,0.07,0.09,0.08) 

plt.figure(figsize=(10,5))

plt.title("Jobs in Market")

plt.axis("equal")

plt.pie(Job_data,labels=Department,explode=explode,colors=colors,autopct='%1.1f%%',startangle=120)

plt.legend()

plt.show()
```
