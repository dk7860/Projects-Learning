#### 2: Write a user-defined function for origin  


```python
#write a user define function
def origin(num):
    if num==1:
        return"USA" 
    elif num==2:
        return"Europe"
    else:
        return"Asia"
#use apply function
df_auto_dataset["origin"]=df_auto_dataset["origin"].apply(origin)
```

#### 3: Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4


```python
sns.pairplot(df_auto_dataset[['mpg','displacement','acceleration','weight','origin']],hue='origin',height=4)
```
